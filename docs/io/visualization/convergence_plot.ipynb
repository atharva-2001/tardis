You only need to include `export_cplots=True` in the `run_tardis` function when you want to share the notebook. The function shows the plot using the Plotly `notebook_connected` renderer, which helps display the plot online. You don't need to do it when running the notebook locally.

```python
from tardis import run_tardis
sim = run_tardis('tardis_example.yml', export_cplots = True)
```

### Displaying Convergence Plots
You can also call the plots outside of `run_tardis` function. 

```python
sim.cplots.plasma_plot.show(renderer="notebook_connected")
```

```python
sim.cplots.luminosity_plot.layout.template = "seaborn"
sim.cplots.luminosity_plot.show(renderer="notebook_connected")
```

You can override the default layout by passing dictionaries as arguments in `luminosity_plot_config` and `plasma_plot_config`  in the `run_tardis` function. 
The default color scales of both the plots can be changed by passing the name of the colorscale in the `plasma_cmap` and the `other_plot_colors` option. One can also provide a list of colors in rgb, hex or css-names format to change the default colors of the luminosity and inner boundary temperature plots. 
Let's change a few things to try that.

```python
sim = run_tardis(
    "tardis_example.yml",
    plasma_plot_config={
        "layout": {
            "template": "plotly_dark",
            "xaxis2": {"title": {"text": "Changed this"}},
        },
    },
    luminosity_plot_config={
        "layout": {
            "template": "plotly_dark"
        },
    },
    plasma_cmap= "viridis", 
    other_plot_colors = ['rgb(102, 197, 204)',
                         'rgb(246, 207, 113)',
                         'rgb(248, 156, 116)',
                         'rgb(220, 176, 242)',
                         'rgb(135, 197, 95)'],
    export_cplots = True
)
```
