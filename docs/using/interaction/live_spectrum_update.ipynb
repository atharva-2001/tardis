# updating spectrum with each iteration

```python
from tardis import run_tardis
import tardis
```

```python
import plotly.graph_objects as go
fig = go.FigureWidget()
fig.add_scatter(x = list(range(500, 20000))[0::80])

fig.update_yaxes( title_text = r"$Luminosity$", range=[0, 7e39])
fig.update_xaxes( title_text = r"$Wavelength$")

fig.update_layout(showlegend = False)
fig
```

```python
fig.show(renderer = "notebook_connected")
```

```python
i = 0
```

```python
def add_scatter(sim):
    """
    adds another scatter trace of Luminosity vs Wavelength with dark blue line color
    """
    fig.add_scatter(x = sim.runner.spectrum.wavelength.value.tolist()[0::80], y = sim.runner.spectrum.luminosity_density_lambda.value.tolist()[0::80], line_color="#0000ff")

def fix_colors(index):
    """
    changes line color of any trace from dark blue to light blue
    """
    fig['data'][index]['line']['color']="#7dafff"
    
def update_graph(sim):
    '''
    updates graphs
    '''
    global i
    i = i+1
    fix_colors(i-1)
    add_scatter(sim)
```

```python
sim = run_tardis('tardis_example.yml', 
                 simulation_callbacks=[[update_graph]])
```
