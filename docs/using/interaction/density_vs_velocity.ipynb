# Density vs Velocity Plot

```python
from tardis import run_tardis
import tardis
```

```python
sim = run_tardis('tardis_example.yml')
```

```python
import plotly.graph_objects as go
fig = go.Figure(data = go.Scatter(y = sim.model.density, x = sim.model.velocity))

fig.update_yaxes(title_text = r"$Density~[g/cm^3]$", showexponent = 'all',exponentformat = 'e')
fig.update_xaxes(title_text = r"$Velocity~[cm/s]$",showexponent = 'all',exponentformat = 'e')

fig.show(renderer = "notebook_connected")
```
